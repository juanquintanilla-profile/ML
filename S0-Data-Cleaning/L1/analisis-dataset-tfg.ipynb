```python
import matplotlib.pyplot as plt
import math

# Agrupamos por enfermedad
groups = list(df.groupby("Enfermedad"))

n = len(groups)
ncols = 3                      # columnas del grid
nrows = math.ceil(n / ncols)   # filas necesarias

fig, axes = plt.subplots(nrows, ncols, figsize=(12, 3 * nrows), sharey=True)
axes = axes.flatten()

ymax = df["Num_toses"].max()   # misma escala para todos

for ax, (enfermedad, subdf) in zip(axes, groups):
    ax.bar(
        subdf["ID"].astype(str),
        subdf["Num_toses"]
    )
    ax.set_title(enfermedad, fontsize=9)
    ax.set_ylim(0, ymax)
    ax.set_xlabel("ID")
    ax.set_ylabel("Nº de toses")

# Eliminamos subplots vacíos si sobran
for ax in axes[len(groups):]:
    ax.axis("off")

fig.suptitle(
    "Contribución de cada paciente al número total de toses por enfermedad",
    fontsize=12
)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,4))

for sexo in df["Sexo"].unique():
    sub = df[df["Sexo"] == sexo]
    plt.scatter(
        sub["Edad"],
        sub["Num_toses"],
        label=sexo,
        alpha=0.7
    )

plt.xlabel("Edad")
plt.ylabel("Número de toses")
plt.title("Relación entre edad, sexo y número de toses")
plt.legend(title="Sexo")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))

for sexo in df["Sexo"].unique():
    sub = df[df["Sexo"] == sexo]
    plt.scatter(
        sub["Edad"],
        sub["Num_toses"],
        s=sub["Num_toses"] / 5,
        alpha=0.6,
        label=sexo
    )

plt.xlabel("Edad")
plt.ylabel("Número de toses")
plt.title("Edad y sexo vs volumen de muestras")
plt.legend(title="Sexo")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear bins de edad (clínicamente razonables)
df["Edad_bin"] = pd.cut(
    df["Edad"],
    bins=[0, 40, 60, 80, 100],
    labels=["<40", "40–60", "60–80", ">80"]
)

# Agregar por sexo y rango de edad (sumando toses)
heat = (
    df.groupby(["Sexo", "Edad_bin"])["Num_toses"]
      .sum()
      .unstack(fill_value=0)
)

plt.figure(figsize=(6,4))
plt.imshow(heat, aspect="auto")
plt.colorbar(label="Número de toses")

plt.xticks(range(len(heat.columns)), heat.columns)
plt.yticks(range(len(heat.index)), heat.index)

plt.xlabel("Rango de edad")
plt.ylabel("Sexo")
plt.title("Distribución de muestras por sexo y rango de edad")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Bins de edad
df["Edad_bin"] = pd.cut(
    df["Edad"],
    bins=[0, 40, 60, 80, 100],
    labels=["<40", "40–60", "60–80", ">80"]
)

# Agregación: TOTAL de toses (muestras)
heat = (
    df.groupby(["Sexo", "Edad_bin"])["Num_toses"]
      .sum()
      .unstack(fill_value=0)
)

plt.figure(figsize=(6,4))
sns.heatmap(
    heat,
    annot=True,
    fmt="d",
    cmap="viridis"
)

plt.xlabel("Rango de edad")
plt.ylabel("Sexo")
plt.title("Distribución de muestras por sexo y rango de edad")
plt.tight_layout()
plt.show()
```
